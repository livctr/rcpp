```python
import numpy as np
import matplotlib.pyplot as plt

# Define y_hat and functions
y_hat = np.arange(0, 1, 0.01)

def type_II_error(thresh, y_hat):
    return np.where(y_hat > 1 - thresh, 0, 1)

def surrogate_type_II_error(thresh, y_hat, tau=0.1, L=5):
    return tau * (1 - thresh) + np.clip(L * (1 - thresh - y_hat), 0, 1)

# Set threshold
thresh = 0.5

# Compute error values for each function
error_typeII = type_II_error(thresh, y_hat)
error_surrogate1 = surrogate_type_II_error(thresh, y_hat, tau=0.1, L=5)
error_surrogate2 = surrogate_type_II_error(thresh, y_hat, tau=0.645, L=5)

# Create the plot with distinct line themes
plt.figure(figsize=(8, 6))
plt.plot(y_hat, error_typeII, color='blue', linestyle='-', linewidth=2, label='Type II Error')
plt.plot(y_hat, error_surrogate1, color='red', linestyle='--', linewidth=2, label=r'Surrogate Loss ($\tau=0.1$)')
plt.plot(y_hat, error_surrogate2, color='green', linestyle=':', linewidth=2, label=r'Surrogate Loss ($\tau=0.645$)')

# Label the axes
plt.xlabel(r'$\hat{y}$')
plt.ylabel('Error')
plt.ylim(-0.1, 1.5)
plt.xlim(0, 1)
plt.legend()
plt.show()
```
